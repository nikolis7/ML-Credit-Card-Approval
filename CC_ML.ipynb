## Define Problem

**Machine Learning Task** 

Credit score cards are a common risk control method in the financial industry. It uses personal information and data submitted by credit card applicants to predict the probability of future defaults and credit card borrowings. The bank is able to decide whether to issue a credit card to the applicant. Credit scores can objectively quantify the magnitude of risk.
 
Generally speaking, credit score cards are based on historical data. Once encountering large economic fluctuations. Past models may lose their original predictive power. Logistic model is a common method for credit scoring. Because Logistic is suitable for binary classification tasks and can calculate the coefficients of each feature. In order to facilitate understanding and operation, the score card will multiply the logistic regression coefficient by a certain value (such as 100) and round it.
 
At present, with the development of machine learning algorithms. More predictive methods such as Boosting, Random Forest, and Support Vector Machines have been introduced into credit card scoring. However, these methods often do not have good transparency. It may be difficult to provide customers and regulators with a reason for rejection or acceptance.

* If the skewness is between **-0.5 and 0.5**, the data are fairly **symmetrical**

* If the distribution is tall and thin it is called a leptokurtic distribution (**Kurtosis > 3**). Values in a **leptokurtic** distribution are near the mean or at the extremes.
* A flat distribution where the values are moderately spread out (i.e., unlike leptokurtic) is called **platykurtic** (**Kurtosis <3**) distribution.
* A distribution whose shape is in between a leptokurtic distribution and a platykurtic distribution is called a mesokurtic (__Kurtosis=3__) distribution. A __mesokurtic__ distribution looks more close to a normal distribution.

**High kurtosis in a data set is an indicator that data has heavy outliers.**


**Low kurtosis in a data set is an indicator that data has lack of outliers.**

##### Missing values & Imputation Strategy 

There are three types of missing values: 

* Missing Completely at Random (MCAR): when data is completely missing at random across the dataset with no discernable pattern. 

* Missing At Random (MAR): when data is not missing randomly, but only within sub-samples of data. 

* Not Missing at Random (NMAR), when there is a noticeable trend in the way data is missing.

For this specific dataset we have MAR since the missing values come only from the feature OCCUPATION_TYPE.

Having a missing value in a machine learning model is considered very inefficient and hazardous because of the following reasons:

* Reduces the efficiency of the ML model.
* Affects the overall distribution of data values.
* It leads to a biased effect in the estimation of the ML model.

Thus we will apply imputation techniques in order to overcome the above problems.

As far as the NaN values we created in DAYS_EMPLOYED we will replace NaN by mean value.

For the imputation of Occupation type in order to avoid adding bias in the data we will choose randomly from the existing values.

```python
import random 

# Creating a list with the possible values
a = df.loc[:,"OCCUPATION_TYPE"].value_counts().index.tolist()

# how many values i want to create
n = df.loc[df.loc[:,"OCCUPATION_TYPE"].isnull(),"OCCUPATION_TYPE"].shape[0]

# Taking the original weights in a tuple
helper = pd.DataFrame(df.loc[:,"OCCUPATION_TYPE"].value_counts())
helper['weight'] = helper.OCCUPATION_TYPE/helper.OCCUPATION_TYPE.sum()

# create a list of n values keeping the weight of original data values 
values=[]
for i in range(n): 
    values.append (random.choices(a,  weights=tuple(helper.weight.tolist())))
```

```python
# DAYS_EMPLOYED imputation by mean

df.loc[df.loc[:,"DAYS_EMPLOYED"].isnull(),"DAYS_EMPLOYED"]=df.loc[:,"DAYS_EMPLOYED"].mean()

# OCCUPATION_TYPE imputation by most frequent value (Laborers) [We are taking the risk to add bias in the model]
df.loc[df.loc[:,"OCCUPATION_TYPE"].isnull(),"OCCUPATION_TYPE"]=values
```

##### Discretization 

Based on our analysis there are extreme values for the features **CNT_FAM_MEMBERS**, **CNT_CHILDREN** and **AMT_INCOME_TOTAL**. We are going to smooth the data by applying discretization. We will also apply to all the rest numerical data discretization in order to smooth the data.   

```python
from sklearn.preprocessing import KBinsDiscretizer

# For AMT_INCOME_TOTAL

# retrieve just the numeric input values
data = df.AMT_INCOME_TOTAL.values.reshape(-1, 1)

# perform k-means discretization (number of clusters = 10)
trans = KBinsDiscretizer(n_bins=10, encode='ordinal', strategy='kmeans')
data_t = trans.fit_transform(data)

# replace the new data
df.loc[:, "AMT_INCOME_TOTAL"] = data_t

df.loc[: , "AMT_INCOME_TOTAL"] = df.loc[:, "AMT_INCOME_TOTAL"].astype(int)



# For CNT_FAM_MEMBERS


# retrieve just the numeric input values
data1= df.loc[:, "CNT_FAM_MEMBERS"].values.reshape(-1, 1)

# perform k-means discretization (number of clusters = 3)
trans = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='kmeans')
data1_t = trans.fit_transform(data1)

# replace the new data
df.loc[:, "CNT_FAM_MEMBERS"] = data1_t

df.loc[:, "CNT_FAM_MEMBERS"] = df.loc[:, "CNT_FAM_MEMBERS"].astype(int)


# For CNT_CHILDREN


# retrieve just the numeric input values
data1= df.CNT_CHILDREN.values.reshape(-1, 1)

# perform k-means discretization (number of clusters = 3)
trans = KBinsDiscretizer(n_bins=3, encode='ordinal', strategy='kmeans')
data1_t = trans.fit_transform(data1)

# replace the new data
df.loc[:, "CNT_CHILDREN"] = data1_t

df.loc[:, "CNT_CHILDREN"] = df.loc[:, "CNT_CHILDREN"].astype(int)
```
